### **1. Trasformare cicli in list comprehension**

Fai tre esempi di cicli producono una lista e poi riscrivi lo stesso codice usando la list comprehension

```python
#Lista di quadrati di numeri fino a 10
quadrati1=[]
for i in range(11):
    quadrati1+=[i**2]
print(quadrati1)
quadrati2=[i**2 for i in range(11)]
print(quadrati2)
#Lista di numeri pari fino a 20
pari1=[]
for i in range(21):
    if i % 2 == 0:
        pari1+=[i]
print(pari1)
pari2=[i for i in range(21) if i % 2 == 0]
print(pari2)

#Lista di numeri primi fino a 50
primi1=[]
for n in range(51):
    if n > 1:
        primo = True
        for i in range(2, n):
            if n % i == 0:
                primo = False
                break
        if primo:
            primi1.append(n)
#inizio a guardare da 2, poi so che gia e divisibile per 1, e range va fino a n-1, poi se trova un numero minore di se stesso che lo divide, diventa false e smette di guardare avanti
print(primi1)
primi2 = [n for n in range(51) if n > 1 and all(n % i != 0 for i in range(2, n))]
print(primi2)
```

### **3.  Operazioni su array**
Crea un array **a**, 1D (senza digitarlo esplicitamente), e genera due nuovi array **b**, contenente una sottostringa a piacere e **c** che è il reverse dell'array **a** (esempio [1,2,3] diventa [3,2,1]) .  Dividi l'array **a** per l'array **c**. Fai la stessa cosa per una lista.

```python
#array di numpy
import numpy as np
# array a, 1D, senza scriverlo esplicitamente
a = np.array(range(1, 6))  
b = a[1:4]
# reverse di a
c = a[::-1]
print("a =", a)
print("b =", b)
print("c =", c)
# divisione
print("a/c=", a/c)

#lista di python
lista_a=list(range(1,6))
lista_b=lista_a[1:4]
lista_c=lista_a[::-1]
print("a =", lista_a)
print("b =", lista_b)
print("c =", lista_c)
#divisione
for i in range(len(lista_a)):
    print(f"{lista_a[i]}/{lista_c[i]}={lista_a[i] / lista_c[i]}")
```
